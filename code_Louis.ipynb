# Crer les jeux de test et d'entranement

Cette section concerne la cration des ensembles de donnes utiliss pour l'apprentissage et les tests. Nous regroupons les donnes relatives aux patients selon leurs identifiants uniques, puis nous les tiquetons en fonction de la prsence ou non d'une condition spcifique. Cette tape permet ainsi la formation de l'ensemble d'apprentissage. De plus, nous slectionnons alatoirement un sous-ensemble de patients pour constituer l'ensemble de test.

Le but de cet tape est la division le jeu de donnes pour laisser entrainer le modle et ainsi le valuer. La division raisonnable est crucial pour obtenir un vrai metrique de modle et aussi viter "overfitting" ou "underfitting".

Puis, le modle s'entrainera sur le jeu d'entrainement et aprs on calcule la metrique sur le jeu de test. Ce metrique montre comment notre modle marche sur les donnes rels et ainsi on peut comparer les modles differentes.

# Fonction pour diviser les documents en plus petits morceaux

Dans cette partie, nous dfinissons une fonction permettant de dcouper les documents textuels en segments plus petits afin de faciliter leur traitement ultrieur.

````python
def split_text(text, k):
    # Convertir le texte en une liste de mots
    words = text.split()

    # Dterminer le nombre total de mots dans le texte
    num_words = len(words)

    # Calculer le nombre de mots par partie
    words_per_part = num_words // k

    # Calculer le nombre de mots restants si num_words n'est pas un multiple de k
    remainder = num_words % k

    # Initialiser une liste pour stocker les parties dcoupes du texte
    parts = []

    # Initialiser l'indice de dbut pour la dcoupe
    start = 0

    # Parcourir chaque partie
    for i in range(k):
        # Calculer la position de fin pour la i-me partie
        end = start + words_per_part + (i < remainder)
        # La variable "end" correspond  la position du dernier mot de la i-me partie

        # Ajouter la partie actuelle  la liste des parties
        parts.append(words[start:end])

        # Mettre  jour l'indice de dbut pour la prochaine partie
        start = end

    # Convertir les listes de mots en chanes de caractres
    parts = [" ".join(part) for part in parts]

    return parts
````

````python
# Importer les bibliothques ncessaires
from transformers import AutoTokenizer, AutoModel
#import torch
from torch import nn

# Charger le modle de langue pr-entran (Bio_ClinicalBERT) et le tokenizer associ
tokenizer = AutoTokenizer.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
model = AutoModel.from_pretrained("emilyalsentzer/Bio_ClinicalBERT").to("cuda")

# Slectionner alatoirement 1000 individus de chaque classe (label_1 et label_0)
sample = pd.concat([label_1.sample(n=5000, random_state=56), label_0.sample(n=30000, random_state=78)]).reset_index().drop('index', axis=1)

# Filtrer le dataframe de donnes en ne conservant que les patients slectionns prcdemment
filtered_data = data[data["SUBJECT_ID"].isin(sample["SUBJECT_ID"].values)]

# Regrouper les donnes filtres par 'SUBJECT_ID' en agrgeant les listes de 'TEXT' et 'TIME'
grouped_sample = filtered_data.groupby('SUBJECT_ID').agg({'TEXT': list, 'TIME': list}).reset_index()
````

## [TRAIN] Extraire les tokens CLS

Dans cette partie, le code traite les donnes textuelles en les divisant en segments plus petits et extrait les embeddings ClinicalBERT correspondants  ces segments. Le rsultat est un dictionnaire qui associe chaque patient  ses embeddings ClinicalBERT.

````python
# Crer un dictionnaire de la forme {n patient: [liste des textes, valeurs time]} pour faciliter l'itration
grouped_texts_dict = grouped_sample.set_index('SUBJECT_ID')[['TEXT', 'TIME']].to_dict(orient='index')

# Initialiser une liste pour stocker les valeurs 'TIME' de chaque partie d'un document
time_list = []

# Initialiser un dictionnaire pour stocker les embeddings
embeddings_dict = {}

# Parcourir les patients et leurs donnes associes
for subject_id, values in grouped_texts_dict.items():
    texts = values['TEXT']  # Rcuprer la liste des documents
    times = values['TIME']  # Rcuprer la liste des valeurs 'TIME' associes aux documents
    embeddings_list = []  # Liste pour stocker les embeddings de toutes les parties de tous les documents

    # Parcourir les documents et leurs valeurs 'TIME' associes
    for text, time in zip(texts, times):
        # Diviser le texte en parties gales
        encoded_text = tokenizer.encode(text)  # Encodage du texte en une squence de tokens
        n_tokens = len(encoded_text)  # Nombre de tokens dans la squence
        n_chunks = max(1, n_tokens // 512)  # Calcul du nombre optimal de parties
        parties = split_text(text, n_chunks)  # Liste des parties du texte

        # Stocker les embeddings des diffrentes parties du document
        cls_embeddings_list = []  # Liste pour stocker les embeddings [CLS] des parties

        # Parcourir les parties du document
        for partie in parties:
            # Convertir la partie dans un format compatible avec le modle
            inputs = tokenizer(partie, return_tensors='pt', padding=True, truncation=True, max_length=512).to("cuda")

            with torch.no_grad():
                # Effectuer l'infrence pour obtenir les rsultats du modle
                outputs = model(**inputs)

            # Rcuprer l'embedding du token [CLS] pour chaque partie
            cls_embeddings = outputs.last_hidden_state[:, 0, :]
            # Stocker l'embedding dans la liste
            cls_embeddings_list.append(cls_embeddings)
            # Stocker la valeur 'TIME' (la mme pour toutes les parties du mme document)
            time_list.append(time)

        # Ajouter la liste des embeddings [CLS]  la liste des embeddings de ce document
        embeddings_list += cls_embeddings_list

    # Stocker les embeddings dans un dictionnaire avec le numro du patient comme cl
    embeddings_dict[subject_id] = torch.stack(embeddings_list)
````

### [TRAIN] ACP

Dans cette section, nous appliquons une Analyse en Composantes Principales (ACP) aux embeddings. L'objectif de l'ACP est de rduire davantage la dimensionnalit des donnes tout en prservant autant d'informations que possible. Le code calcule la variance explique par chaque composante principale, ce qui permet d'valuer l'efficacit de la rduction de dimension.

````python
# Importer la classe PCA (Analyse en Composantes Principales) du module sklearn
from sklearn.decomposition import PCA

# Crer un nouveau dictionnaire pour stocker les embeddings aplatis
flattened_embeddings_dict = {key: tensor.view(tensor.shape[0], -1).cpu().numpy() for key, tensor in embeddings_dict.items()}

# Crer une liste pour stocker le nombre d'embeddings pour chaque patient
lengths = [len(tensor) for tensor in flattened_embeddings_dict.values()]

# Concatner tous les embeddings en une seule matrice
embeddings_np = np.concatenate(list(flattened_embeddings_dict.values()))

# Initialiser le modle PCA (Analyse en Composantes Principales) avec 100 composantes
pca = PCA(n_components=100)

# Ajuster le modle PCA aux donnes et les transformer pour rduire la dimension
reduced_embeddings_np = pca.fit_transform(embeddings_np)

# Sparer les embeddings transforms pour chaque sujet
reduced_embeddings_list = np.split(reduced_embeddings_np, np.cumsum(lengths)[:-1])

# Recrer le dictionnaire des embeddings transforms avec les cls correspondantes
reduced_embeddings_dict = {key: tensor for key, tensor in zip(flattened_embeddings_dict.keys(), reduced_embeddings_list)}

# Filtrer les embeddings dont la forme est diffrente de (1, 100)
filtered_embeddings_dict = {key: value for key, value in reduced_embeddings_dict.items() if value.shape != (1, 100)}
reduced_embeddings_dict = filtered_embeddings_dict
del filtered_embeddings_dict  # Supprimer la variable temporaire

# Ajouter la valeur 'time'  la fin de chaque embedding
for key, time in zip(reduced_embeddings_dict.keys(), time_list):
    # Crer un tableau vide pour stocker les embeddings avec le temps
    array_vide = np.empty((reduced_embeddings_dict[key].shape[0], 101))
    for i in range(reduced_embeddings_dict[key].shape[0]):
        # Ajouter la valeur 'time'  la fin de chaque embedding
        array = np.append(reduced_embeddings_dict[key][i], time)
        array_vide[i:] = array
    reduced_embeddings_dict[key] = array_vide

# Afficher la variance explique par chaque composante principale
print("Variance explique par chaque composante principale:", pca.explained_variance_ratio_)

# Afficher la variance totale explique par toutes les composantes principales
print("Variance totale explique:", sum(pca.explained_variance_ratio_))

# Vrifier que le lien entre chaque sujet et ses embeddings est bien conserv
for key in reduced_embeddings_dict.keys():
    print(f"ID du patient : {key}, Forme des embeddings : {reduced_embeddings_dict[key].shape}")
    break
````

## [TRAIN] Calculer les signatures

Cette partie du code calcule les signatures logarithmiques pour les embeddings rduits.
Les signatures logarithmiques capturent des informations plus complexes dans les donnes, ce qui peut tre trs utile lors de l'entranement d'un modle de prdiction. Cela aboutit  la cration d'un dictionnaire o chaque patient est reprsent par des plongements sous forme de signatures logarithmiques.

````python
# Importer les bibliothques ncessaires
#import torch
import signatory

# Ordre de la signature tronque
depth = 2

# Crer un nouveau dictionnaire pour stocker les rsultats de la log signature
log_signature_dict = {}

# Parcourir le dictionnaire des embeddings rduits (reduced_embeddings_dict)
for key, value in reduced_embeddings_dict.items():
    # Convertir les tableaux NumPy en tenseurs PyTorch de type float
    tensor = torch.from_numpy(value).float().to("cuda")

    # Ajouter une dimension "batch" pour correspondre au format requis (batch, stream, channel)
    tensor = tensor.unsqueeze(0).to("cuda")

    # Calculer la log signature en utilisant la bibliothque Signatory
    log_signature = signatory.logsignature(path=tensor, depth=depth)

    # Enlever la dimension "batch" que nous avons ajoute prcdemment
    log_signature = log_signature.squeeze(0).to("cuda")

    # Ajouter le rsultat dans le dictionnaire log_signature_dict
    log_signature_dict[key] = log_signature

#  ce stade, log_signature_dict est un dictionnaire o chaque cl correspond  un numro de patient, et chaque valeur est la log signature de ce patient.
````

````python
from sklearn.decomposition import PCA

# Crer un nouveau dictionnaire pour stocker les embeddings aplatis
flattened_embeddings_dict_test = {key: tensor.view(tensor.shape[0], -1).cpu().numpy() for key, tensor in embeddings_dict_test.items()}

# Crer une liste pour stocker le nombre d'embeddings pour chaque patient
lengths = [len(tensor) for tensor in flattened_embeddings_dict_test.values()]

# Concatner tous les embeddings dans une seule matrice numpy
embeddings_np_test = np.concatenate(list(flattened_embeddings_dict_test.values()))

# Initialiser l'analyse en composantes principales (ACP) avec 100 composantes
pca = PCA(n_components=100)

# Ajuster le modle ACP sur les donnes et les transformer
reduced_embeddings_np_test = pca.fit_transform(embeddings_np_test)

# Sparer les embeddings transforms pour chaque sujet
reduced_embeddings_list_test = np.split(reduced_embeddings_np_test, np.cumsum(lengths)[:-1])

# Recrer le dictionnaire des embeddings transforms avec les cls correspondantes
reduced_embeddings_dict_test = {key: tensor for key, tensor in zip(flattened_embeddings_dict_test.keys(), reduced_embeddings_list_test)}

# Filtrer les embeddings pour exclure ceux de forme (1, 100)
filtered_embeddings_dict_test = {key: value for key, value in reduced_embeddings_dict_test.items() if value.shape != (1, 100)}
reduced_embeddings_dict_test = filtered_embeddings_dict_test
del filtered_embeddings_dict_test

# Ajouter la valeur 'time'  la fin de chaque embedding rduit
for key, time in zip(reduced_embeddings_dict_test.keys(), time_list_test):
    # Crer un tableau vide pour stocker les embeddings et le temps
    array_vide = np.empty((reduced_embeddings_dict_test[key].shape[0], 101))
    for i in range(reduced_embeddings_dict_test[key].shape[0]):
        # Concatner l'embedding rduit avec la valeur 'time'
        array = np.append(reduced_embeddings_dict_test[key][i], time)
        array_vide[i:] = array
    reduced_embeddings_dict_test[key] = array_vide
    
temps = [i for i in range(0,100)]

import matplotlib.pyplot as plt

plt.plot(temps, pca.explained_variance_ratio_)
plt.xlabel('X-axis Label')  # Add x-axis label
plt.ylabel('Y-axis Label')  # Add y-axis label
plt.title('Title of the Plot')  # Add title
plt.show()

# Afficher la variance explique par chaque composante principale
print("Variance explained by each component:", pca.explained_variance_ratio_)
print("Total variance explained:", sum(pca.explained_variance_ratio_))

# Vrifier que le lien entre chaque sujet et ses embeddings est bien conserv
for key in reduced_embeddings_dict_test.keys():
    print(f"Subject ID: {key}, Embedding shape: {reduced_embeddings_dict_test[key].shape}")
    break
````

## [TEST] Calculer les signatures

Ici, le programme calcule les signatures logarithmiques pour les embeddings des donnes de test. Ces signatures contiennent des informations plus avances qui peuvent tre bnfiques pour l'entranement du modle. Le rsultat est un dictionnaire o les embeddings de chaque patient sont reprsents sous forme de signatures logarithmiques.

````python
#import torch
import signatory

# Ordre de la signature tronqu
depth = 2

# Crer un nouveau dictionnaire pour stocker les rsultats de la log signature
log_signature_dict_test = {}

# Parcourir le dictionnaire des embeddings rduits pour chaque patient
for key, value in reduced_embeddings_dict_test.items():
    # Convertir le numpy array en un torch tensor et spcifier le type comme float
    tensor = torch.from_numpy(value).float()
    # Alternative : utiliser directement torch.tensor(value, dtype=torch.float)

    # Ajouter une dimension supplmentaire pour correspondre  l'exigence du modle (batch, stream, channel)
    tensor = tensor.unsqueeze(0)

    # Calculer la log signature en utilisant le modle signatory
    log_signature = signatory.logsignature(path=tensor, depth=depth)

    # Enlever la dimension batch que nous avons ajoute prcdemment
    log_signature = log_signature.squeeze(0)

    # Ajouter le rsultat au dictionnaire avec le numro du patient comme cl
    log_signature_dict_test[key] = log_signature

# Maintenant, log_signature_dict est un dictionnaire o chaque cl correspond  un patient et chaque valeur est la log signature de ce sujet
````

Les performances du modle sont values selon la prcision, le rappel et le score F1.

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, recall_score, f1_score
import math

# Sparation des donnes en ensembles d'entranement et de test

# Slectionner toutes les colonnes sauf la dernire (la colonne 'HOSPITAL_EXPIRE_FLAG') comme caractristiques d'entranement
X_train = df_final.iloc[:, 1:-1]

# Slectionner la dernire colonne ('HOSPITAL_EXPIRE_FLAG') comme variable cible d'entranement et la convertir en entiers
y_train = df_final["HOSPITAL_EXPIRE_FLAG"].astype(int)

# Slectionner les caractristiques de l'ensemble de test de manire similaire
X_test = df_final_test.iloc[:, 1:-1]
y_test = df_final_test["HOSPITAL_EXPIRE_FLAG"].astype(int)

# Dfinir les valeurs des hyperparamtres  tester

# Crer un dictionnaire de valeurs  tester pour le nombre maximal de caractristiques utilises  chaque division de l'arbre
param_grid = {
    'max_features': list(range(30, 121, 10))
}

# Initialisation du modle Random Forest

# Crer une instance du modle Random Forest avec des hyperparamtres spcifis
model = RandomForestClassifier(criterion='entropy', n_estimators=150)

# Recherche des meilleurs hyperparamtres

# Crer un objet GridSearchCV pour effectuer une recherche des meilleurs hyperparamtres
# cv=5 indique une validation croise en 5 plis et scoring='f1_micro' utilise le F1-score pour l'valuation
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='f1_micro')

# Effectuer la recherche des meilleurs hyperparamtres en utilisant les donnes d'entranement
grid_search.fit(X_train, y_train)

# Obtenir les meilleures valeurs des hyperparamtres
best_max_features = grid_search.best_params_['max_features']

# Utilisation du modle avec les meilleurs hyperparamtres pour la prdiction sur l'ensemble de test

# Crer une nouvelle instance du modle Random Forest avec les meilleurs hyperparamtres trouvs
best_model = RandomForestClassifier(criterion='entropy', n_estimators=150, max_features=best_max_features)

# Entraner le modle avec les meilleurs hyperparamtres sur les donnes d'entranement
best_model.fit(X_train, y_train)

# Prdire les tiquettes sur l'ensemble de test en utilisant le modle entran
y_pred = best_model.predict(X_test)

# valuation des performances du modle

# Calculer et afficher l'accuracy du modle
accuracy = accuracy_score(y_test, y_pred)
print("Prcision: %.2f%%" % (accuracy * 100.0))

# Calculer et afficher le rappel du modle
recall = recall_score(y_test, y_pred)
print("Rappel: %.2f%%" % (recall * 100.0))

# Calculer et afficher le F1-score du modle
f1 = f1_score(y_test, y_pred)
print("F1-score: %.2f%%" % (f1 * 100.0))
````
